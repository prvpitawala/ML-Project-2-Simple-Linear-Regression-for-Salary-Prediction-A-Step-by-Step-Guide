# **Simple Linear Regression**

This notebook demonstrates how to build a simple linear regression model using Python's scikit-learn library to predict an employee's salary based on their years of experience. The notebook provides a step-by-step guide through the entire process, from data preprocessing to model evaluation and visualization.

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
# Get the dataset
dataset = pd.read_csv('Salary_Data.csv')
print(dataset[:10])
```

```python
# divide in to Independent variables and the dependent variables
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1:].values
```

```python
# Indipendent variables
print(X)
```

```python
# Dependent variables
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
# import Train test split
from sklearn.model_selection import train_test_split

# Split the dataset in to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

**Simple Linear Regression** is a statistical method that allows us to summarize and study relationships between two continuous (quantitative) variables: an independent variable (predictor) and a dependent variable (response). In this example, the independent variable is "Years of Experience," and the dependent variable is "Salary." The goal of Simple Linear Regression is to create a model that can predict the dependent variable based on the independent variable.

```python
# Import the simple LinearRegression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# Fit the model to X train and y train
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
# Get the predictions
y_pred = regressor.predict(X_test)
```

```python
# Print the predictions
print(y_pred)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color='red', label='Actual Salary (train set)')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Regression Line')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color='red', label='Actual Salary (Test set)')
plt.plot(X_train, regressor.predict(X_train), color='blue', label='Regression Line (Training set)')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
